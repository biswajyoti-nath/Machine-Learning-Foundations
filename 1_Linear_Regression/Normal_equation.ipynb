```python
from sklearn.datasets import fetch_california_housing

california_housing = fetch_california_housing(as_frame=True)
```

```python
import matplotlib.pyplot as plt

california_housing.frame.hist(figsize=(12, 10), bins=30, edgecolor="black")
plt.subplots_adjust(hspace=0.7, wspace=0.4)
```

```python
features_of_interest = ["AveRooms", "AveBedrms", "AveOccup", "Population"]
california_housing.frame[features_of_interest].describe()
```

```python
import seaborn as sns
sns.scatterplot(
    data=california_housing.frame,
    x="Longitude",
    y="Latitude",
    size="MedHouseVal",
    hue="MedHouseVal",
    palette="viridis",
    alpha=0.5,)
plt.legend(title="MedHouseVal", bbox_to_anchor=(1.05, 0.95), loc="upper left")
_ = plt.title("Median house value depending of\n their spatial location")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
feature = california_housing.data
target = california_housing.target
X_train, X_test, y_train, y_test = train_test_split(
    feature, target, test_size=0.2, random_state=42)
```

```python
mean = X_train.mean()
std = X_train.std() //'''We use the same mean and std to normalize the test set'''
X_train = (X_train - mean) / std //'''We normalize the training set'''
X_test = (X_test - mean) / std //'''We normalize the test set using the same mean and std as the training set'''
```
